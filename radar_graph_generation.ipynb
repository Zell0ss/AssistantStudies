```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definir las categorías para el eje radial
sabores = ['Acidez', "Tanicidad", "Final", 'Fruta','Madera', 'Cuerpo']
values = [5,4,5,4,3,2]

# Crear un gráfico de radar
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)

# Crear los ángulos
angles = np.linspace(0, 2 * np.pi, len(sabores), endpoint=False).tolist()

# Añadir la primera categoría al final para cerrar el círculo
values += values[:1]
angles += angles[:1]

# Dibujar el gráfico de radar
ax.fill(angles, values, color='red', alpha=0.5)


# Colocar las etiquetas en los ángulos adecuados
ax.set_xticks(angles[:-1])
ax.set_xticklabels(sabores)
plt.savefig('radar_graph_generation.jpg')
```

```python
# Definir las categorías para el eje radial
aromas = ['Lácticos', 'Fermentación (Panadería)', 'Yogur', 'Mineral', 'Barrica/coco/vainilla/caramelo', "Frutal (roja y cítrico)"]
values = [0, 2, 2, 0, 1, 3]

# Crear un gráfico de radar
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)

# Crear los ángulos
angles = np.linspace(0, 2 * np.pi, len(aromas), endpoint=False).tolist()

# Añadir la primera categoría al final para cerrar el círculo
values += values[:1]
angles += angles[:1]

# Dibujar el gráfico de radar
ax.fill(angles, values, color='skyblue', alpha=0.5)

# Colocar las etiquetas en los ángulos adecuados
ax.set_xticks(angles[:-1])
ax.set_xticklabels(aromas)
```

```python
# Guardar el gráfico como un archivo jpg
plt.savefig('grafico_aromas.jpg')
```
