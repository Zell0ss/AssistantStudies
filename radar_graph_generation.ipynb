```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definir las categoras para el eje radial
sabores = ['Acidez', "Tanicidad", "Final", 'Fruta','Madera', 'Cuerpo']
values = [5,4,5,4,3,2]

# Crear un grfico de radar
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)

# Crear los ngulos
angles = np.linspace(0, 2 * np.pi, len(sabores), endpoint=False).tolist()

# Aadir la primera categora al final para cerrar el crculo
values += values[:1]
angles += angles[:1]

# Dibujar el grfico de radar
ax.fill(angles, values, color='red', alpha=0.5)


# Colocar las etiquetas en los ngulos adecuados
ax.set_xticks(angles[:-1])
ax.set_xticklabels(sabores)
plt.savefig('radar_graph_generation.jpg')
```

```python
# Definir las categoras para el eje radial
aromas = ['Lcticos', 'Fermentacin (Panadera)', 'Yogur', 'Mineral', 'Barrica/coco/vainilla/caramelo', "Frutal (roja y ctrico)"]
values = [0, 2, 2, 0, 1, 3]

# Crear un grfico de radar
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)

# Crear los ngulos
angles = np.linspace(0, 2 * np.pi, len(aromas), endpoint=False).tolist()

# Aadir la primera categora al final para cerrar el crculo
values += values[:1]
angles += angles[:1]

# Dibujar el grfico de radar
ax.fill(angles, values, color='skyblue', alpha=0.5)

# Colocar las etiquetas en los ngulos adecuados
ax.set_xticks(angles[:-1])
ax.set_xticklabels(aromas)
```

```python
# Guardar el grfico como un archivo jpg
plt.savefig('grafico_aromas.jpg')
```
